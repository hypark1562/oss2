```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 깨짐 방지 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 로드
df = pd.read_csv("../data/processed/cleaned_data.csv")
print(f"데이터 크기: {df.shape}")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['leaguePoints'], bins=30, kde=True, color='skyblue')
plt.title("Challenger League Points Distribution (챌린저 점수 분포)")
plt.xlabel("Points (점수)")
plt.ylabel("Count (인원)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['wins'], y=df['losses'], hue=df['leaguePoints'], palette='viridis')
plt.title("Wins vs Losses Correlation")
plt.show()
```
